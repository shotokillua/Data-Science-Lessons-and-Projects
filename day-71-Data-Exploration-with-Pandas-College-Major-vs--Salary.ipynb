Find the college major with the highest starting salaries. Use the .max() method to find this.

```python
clean_df['Starting Median Salary'].max()
```

So now we know the highest starting salary is 74,300. BUT how do we know which major this value is associated with? You have to use the .idxmax() method

```python
clean_df['Starting Median Salary'].idxmax()
```

Now that we know the index of the highest starting salary, we access the undergraduate major at that corresponding index as shown below:

```python
clean_df['Undergraduate Major'][43] #OR clean_df['undergraduate Major'].loc[43] will produce the same output
```

You can use the .loc property to retrieve an entire row

```python
clean_df.loc[43]
```

```python
spread_col = clean_df['Mid-Career 90th Percentile Salary'].subtract(clean_df['Mid-Career 10th Percentile Salary'])
```

```python
clean_df.insert(1, 'Spread', spread_col)
```

```python
clean_df.head()
```

```python
low_risk = clean_df.sort_values('Spread', ascending=False)
```

```python
low_risk[['Undergraduate Major', 'Spread']].head()
```

**Using the .sort_values() method, can you find the degrees with the highest potential? Find the top 5 degrees with the highest values in the 90th percentile. **

```python
potential = clean_df.sort_values('Mid-Career 90th Percentile Salary', ascending=False)
```

```python
potential[['Undergraduate Major', 'Mid-Career 90th Percentile Salary']].head()
```

**Top 5 degrees w higheest potential are economics, finance, chemical engineering, math, and physics.**

**Top 5 degrees w greatest spread in salaries are: economics, finance, math, marketing, and philosophy.**
