# Data Visualisation

```python
plt.figure(figsize=(14,8), dpi=120)
plt.title('Bitcoin News Search vs Resampled Price')

plt.xticks(fontsize=14, rotation=45)
plt.yticks(fontsize=14)

years = mdates.YearLocator() # added this to get the tick marks
months = mdates.MonthLocator() # added this to get the tick marks
years_fmt = mdates.DateFormatter('%Y') # added this to get the tick marks

ax1 = plt.gca()
ax2 = ax1.twinx()

ax1.plot(df_btc_price_monthly.index, df_btc_price_monthly.CLOSE, color='orange', linewidth=3, linestyle='dashed')
ax2.plot(df_btc_search.MONTH, df_btc_search.BTC_NEWS_SEARCH, color='skyblue', linewidth=3, marker="o")
ax1.set_ylabel('BTC Price ($)', color='orange')
ax2.set_ylabel('Search Trend', color='skyblue')
ax1.set_ylim([0, 15000])
ax2.set_xlim([df_btc_price_monthly.index.min(), df_btc_price_monthly.index.max()])


ax1.xaxis.set_major_locator(years) # added this to get the tick marks
ax1.xaxis.set_major_formatter(years_fmt) # added this to get the tick marks
ax1.xaxis.set_minor_locator(months) # added this to get the tick marks

plt.show()
```

**Big increases in the search trend did not always result from large increases in the stock price, especially after 2018.**

### Unemployement Benefits Search vs. Actual Unemployment in the U.S.

**Challenge** Plot the search for "unemployment benefits" against the unemployment rate.

```python
plt.figure(figsize=(14,8), dpi=120)
plt.title('Monthly Search of "Unemployment Benefits" in the U.S. vs the U/E Rate')

plt.xticks(fontsize=14, rotation=45)
plt.yticks(fontsize=14)

years = mdates.YearLocator() # added this to get the tick marks
months = mdates.MonthLocator() # added this to get the tick marks
years_fmt = mdates.DateFormatter('%Y') # added this to get the tick marks

ax1 = plt.gca()
ax2 = ax1.twinx()

ax1.plot(df_unemployment.MONTH, df_unemployment.UNRATE, color='purple', linewidth=3, linestyle='dashed')
ax2.plot(df_unemployment.MONTH, df_unemployment.UE_BENEFITS_WEB_SEARCH, color='skyblue', linewidth=3)
ax1.set_ylabel('FRED U/E Rate', color='purple')
ax2.set_ylabel('Search Trend', color='skyblue')
ax1.set_ylim([df_unemployment.UNRATE.min()-0.5,df_unemployment.UNRATE.max()+0.5])
ax2.set_xlim([df_unemployment.MONTH.min(), df_unemployment.MONTH.max()])


ax1.xaxis.set_major_locator(years) # added this to get the tick marks
ax1.xaxis.set_major_formatter(years_fmt) # added this to get the tick marks
ax1.xaxis.set_minor_locator(months) # added this to get the tick marks

ax1.grid(color='grey', linestyle="--")

plt.show()
```

**Challenge**: Calculate the 3-month or 6-month rolling average for the web searches. Plot the 6-month rolling average search data against the actual unemployment. What do you see in the chart? Which line moves first?


```python
rolling_df = df_unemployment[['UE_BENEFITS_WEB_SEARCH', 'UNRATE']].rolling(window=6).mean()

plt.figure(figsize=(14,8), dpi=120)
plt.title('Monthly Search of "Unemployment Benefits" in the U.S. vs the U/E Rate')

plt.xticks(fontsize=14, rotation=45)
plt.yticks(fontsize=14)

years = mdates.YearLocator() # added this to get the tick marks
months = mdates.MonthLocator() # added this to get the tick marks
years_fmt = mdates.DateFormatter('%Y') # added this to get the tick marks

ax1 = plt.gca()
ax2 = ax1.twinx()

ax1.plot(df_unemployment.MONTH, rolling_df.UNRATE, color='purple', linewidth=3, linestyle='dashed')
ax2.plot(df_unemployment.MONTH, rolling_df.UE_BENEFITS_WEB_SEARCH, color='skyblue', linewidth=3)
ax1.set_ylabel('FRED U/E Rate', color='purple')
ax2.set_ylabel('Search Trend', color='skyblue')
ax1.set_ylim(bottom=3, top=10.5)
ax2.set_xlim([df_unemployment.MONTH[0], df_unemployment.MONTH.max()])


ax1.xaxis.set_major_locator(years) # added this to get the tick marks
ax1.xaxis.set_major_formatter(years_fmt) # added this to get the tick marks
ax1.xaxis.set_minor_locator(months) # added this to get the tick marks


plt.show()
```
